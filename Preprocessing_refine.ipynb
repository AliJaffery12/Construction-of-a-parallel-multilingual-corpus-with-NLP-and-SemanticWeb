```python
block_dict = {}
page_num = 1
for page in doc: # Iterate all pages in the document
    file_dict = page.get_text('dict') # Get the page dictionary
    block = file_dict['blocks'] # Get the block information
    block_dict[page_num] = block # Store in block dictionary
    page_num += 1 # Increase the page value by 1
```

```python
import re
import pandas as pd
import numpy as np
spans = pd.DataFrame(columns=['xmin', 'ymin', 'xmax', 'ymax', 'text', 'tag'])
rows = []
for page_num, blocks in block_dict.items():
    for block in blocks:
        if block['type'] == 0:
            for line in block['lines']:
                for span in line['spans']:
                    xmin, ymin, xmax, ymax = list(span['bbox'])
                    font_size = span['size']
                    text = unidecode(span['text'])
                    span_font = span['font']
                    is_upper = False
                    is_bold = False
                    if "bold" in span_font.lower():
                        is_bold = True
                    if re.sub("[\(\[].*?[\)\]]", "", text).isupper():
                        is_upper = True
                    if text.replace(" ","") !=  "":
                        rows.append((xmin, ymin, xmax, ymax, text,                              is_upper, is_bold, span_font, font_size))
                        span_df = pd.DataFrame(rows, columns=['xmin','ymin','xmax','ymax', 'text', 'is_upper','is_bold','span_font', 'font_size'])
```

```python
import re
import pandas as pd



# Initialize the hierarchical segmentation
hierarchy = {'name': 'document', 'content': '', 'children': []}
current_chapter = None
current_section = None
current_subsection = None

# Loop through each row in the text_df data frame
for index, row in text_df.iterrows():

    # Extract the heading and content
    heading = row['heading']
    content = row['content']

    # Use regular expressions to identify chapters, sections, and subsections based on their headings
    if heading.startswith('Chapter'):
        # This is a new chapter
        # Append the previous chapter's sections and subsections to the chapters list
        if current_chapter:
            hierarchy['children'].append(current_chapter)
        current_chapter = {'name': heading, 'content': content, 'children': []}
        current_section = None
        current_subsection = None
    elif re.match(r'^\d+\.\d+ .*$', heading):
        # This is a new section
        # Append the previous section's subsections to the sections list
        if current_section:
            current_chapter['children'].append(current_section)
        current_section = {'name': heading, 'content': content, 'children': []}
        current_subsection = None
    elif re.match(r'^\d+\.\d+\.\d+ .*$', heading):
        # This is a new subsection
        current_subsection = {'name': heading, 'content': content}
        current_section['children'].append(current_subsection)
    else:
        # This is content that doesn't have a heading
        if current_subsection:
            current_subsection['content'] += content
        elif current_section:
            current_section['content'] += content
        elif current_chapter:
            current_chapter['content'] += content
        else:
            hierarchy['content'] += content

# Append the last chapter's sections and subsections to the chapters list
if current_chapter:
    hierarchy['children'].append(current_chapter)
if current_section:
    current_chapter['children'].append(current_section)

# Print the hierarchical segmentation
print(hierarchy)
hierarchy_df = pd.DataFrame.from_dict(hierarchy, orient='index')
print(hierarchy_df)
```

```python


# Step 2: Define detect_language function
from langdetect import detect

def detect_language(text):
    try:
        lang = detect(text)
    except:
        lang = 'unknown'
    return lang

# Step 3: Define add_language function
def add_language(segment):
    if 'content' in segment:
        lang = detect_language(segment['content'])
        segment['lang'] = lang
    if 'children' in segment:
        for child in segment['children']:
            add_language(child)

# Step 4: Call add_language on the top-level segment of the hierarchy
add_language(hierarchy)

# Print the updated hierarchy
print(hierarchy)
```

```python
import re
from langdetect import detect

# Concatenate the heading and content columns into a single column
text_column = text_df['heading'] + ' ' + text_df['content']

# Initialize the hierarchical segmentation
chapters = []
sections = []
subsections = []

# Loop through each row in the text_df data frame
for index, row in text_df.iterrows():

    # Extract the heading and content
    heading = row['heading']
    content = row['content']

    # Use regular expressions to identify chapters, sections, and subsections based on their headings
    if heading.startswith('Chapter'):
        # This is a new chapter
        # Append the previous chapter's sections and subsections to the chapters list
        if sections:
            chapters[-1]['sections'] = sections
            sections = []
        chapters.append({'name': heading, 'content': content, 'sections': []})
    elif re.match(r'^\d+\.\d+ .*$', heading):
        # This is a new section
        # Append the previous section's subsections to the sections list
        if subsections:
            sections[-1]['subsections'] = subsections
            subsections = []
        sections.append({'name': heading, 'content': content, 'subsections': []})
    elif re.match(r'^\d+\.\d+\.\d+ .*$', heading):
        # This is a new subsection
        subsections.append({'name': heading, 'content': content})

# Append the last chapter's sections and subsections to the chapters list
if sections:
    chapters[-1]['sections'] = sections
if subsections:
    sections[-1]['subsections'] = subsections

# Initialize a dictionary to store the segment groups by language
segment_groups = {}

# Loop through the chapters and detect the language of each one
for chapter in chapters:
    content = chapter['content']
    # Check that the content is not empty before detecting the language
    if content.strip():
        chapter_lang = detect(content)
        if chapter_lang not in segment_groups:
            segment_groups[chapter_lang] = []
        segment_groups[chapter_lang].append(chapter)

# Print the hierarchical segmentation for each language
for lang, chapter_list in segment_groups.items():
    print(f"Chapters in {lang}:")
    for chapter in chapter_list:
        print(chapter['name'])
        print(chapter['content'])
        for section in chapter['sections']:
            print('\t' + section['name'])
            print('\t' + section['content'])
            for subsection in section['subsections']:
                print('\t\t' + subsection['name'])
                print('\t\t' + subsection['content'])
```

```python
import json
from nltk.corpus.reader import PlaintextCorpusReader
import os

# Load the segmented_data.json file
with open('/content/segmented_data.json', 'r') as f:
    data = json.load(f)

# Convert the data into a list of dictionaries
documents = []
for i, chapter in enumerate(data['en']):
    for j, section in enumerate(chapter['sections']):
        document = {
            'fileid': f'{i + 1}.{j + 1}',
            'raw': section['content'],
            'title': section['name'],
            'chapter': i + 1
        }
        documents.append(document)

# Save the documents to disk
corpus_root = './content'
if not os.path.exists(corpus_root):
    os.makedirs(corpus_root)

for document in documents:
    with open(f'{corpus_root}/{document["fileid"]}.txt', 'w') as f:
        f.write(document['raw'])
        print(f'Wrote document {document["fileid"]}.txt to disk.')

# Check if corpus_root directory exists
if os.path.exists(corpus_root):
    print(f'Corpus directory {corpus_root} exists.')
else:
    print(f'Corpus directory {corpus_root} does not exist.')

# Create a corpus reader object
corpus = PlaintextCorpusReader(corpus_root, '.*\.txt')

# Now you can use the NLTK corpus API to access the corpus data


print(os.getcwd())
```
